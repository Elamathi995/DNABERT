```python
# ============================================================
#  DNABERT
# ============================================================

# 🔹 Step 1: Install libraries (run once in Colab)
# !pip install transformers torch

# 🔹 Step 2: Import Libraries
import torch
import torch.nn as nn
from transformers import AutoTokenizer, AutoModel

# 🔹 Step 3: Load REAL DNABERT model from HuggingFace
model_name = "zhihan1996/DNA_bert_6"

print("🔄 Loading DNABERT model...")

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

print("✅ Model loaded successfully!")

# 🔹 Step 4: Input DNA Sequence
sequence = "ATGCGTACGTTAGC"

# 🔹 Step 5: Convert DNA → k-mers
def kmers(seq, k=6):
    return " ".join([seq[i:i+k] for i in range(len(seq) - k + 1)])

kmer_seq = kmers(sequence, k=6)

print("\n🔹 Original Sequence:")
print(sequence)

print("\n🔹 K-mer Sequence:")
print(kmer_seq)

# 🔹 Step 6: Tokenization
inputs = tokenizer(kmer_seq, return_tensors="pt", padding=True, truncation=True)

# 🔹 Step 7: Run Model
with torch.no_grad():
    outputs = model(**inputs)

# 🔹 Step 8: Extract embeddings
embeddings = outputs.last_hidden_state

print("\n🔹 Embedding Shape:")
print(embeddings.shape)

# 🔹 Step 9: Simple Classification Layer
classifier = nn.Linear(embeddings.shape[-1], 2)

cls_embedding = embeddings[:, 0, :]  # CLS token
logits = classifier(cls_embedding)

print("\n🔹 Logits:")
print(logits)

# 🔹 Step 10: Convert to probabilities
probabilities = torch.softmax(logits, dim=1)

print("\n🔹 Probabilities:")
print(probabilities)

# 🔹 Step 11: Prediction
predicted_class = torch.argmax(probabilities, dim=1)

print("\n🔹 Predicted Class:")
print(predicted_class.item())

print("\n✅ DNABERT pipeline executed successfully!")
```
